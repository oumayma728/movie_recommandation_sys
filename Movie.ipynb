```python
import numpy as np
import pandas as pd
import difflib #to compare texts
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

data collection and pre_processing

```python
df=pd.read_csv('./movies.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#selecting the relevant features for recommandation
selected_features=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
    df[feature] = df[feature].fillna('')
```

```python
df.isnull().sum()
```

```python
#combining all the 5 features
features=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+df['director']
```

```python
#converting the text data to feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(features)
```

Cosine similarity

```python
#getting similarity score
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
movie_input=input('enter your favourite movie name')
```

```python
#creating a list of recommanded movies
r_movies= df['title'].tolist()
#finding the closet match
match=difflib.get_close_matches(movie_input,r_movies)
print(match)
```

```python
close_match=match[0]
```

```python
#index movie
index_movie=df[df['title'] == close_match]['index'].values[0]
print(index_movie)
```

```python
#getting a list of similar movies 
similarity_score = list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of movie 
print("the movies are \n")
i=1 
for movie in sorted_similar_movies:
    index=movie[0]
    title_index=df[df['index'] == index]['title'].values[0]
    if (i<10):
        print(i,'.',title_index)
        i=i+1
```
